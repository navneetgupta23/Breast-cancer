```python
from sklearn.metrics import f1_score,confusion_matrix,classification_report
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_stan = pd.concat([data_stan, df['target']], axis=1)
data_stan.head()
```

```python
data_stan['target'] = data_stan['target'].replace({'M':0, 'B':1})
data_stan.head()
```

```python
corr = data_stan.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(23, 13))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, annot=True,fmt='.2f',mask=mask, cmap=cmap, ax=ax);
```

```python
X = data_stan.drop('target', axis=1)
y=data_stan['target']
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
params={
    'criterion': ['gini','entropy'],
    'max_depth': range(1,16,2),
    'min_samples_split': range(1,16,2),
    'max_features': range(1,16,2)
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
tree_cv = RandomizedSearchCV(dt, params, cv = 5,return_train_score=False )
tree_cv.fit(X, y) 
```

```python
print("Tuned Decision Tree Parameters: {}".format(tree_cv.best_params_)) 
print("Best score is {}".format(tree_cv.best_score_))
```

```python
from sklearn.model_selection import GridSearchCV
g_search = GridSearchCV(estimator = dt, param_grid = params, cv = 5, n_jobs = -1, verbose = 0, return_train_score=True)
```

```python
g_search.fit(X_train, y_train)
```

```python
print(g_search.best_params_)
```

```python
dt1 = DecisionTreeClassifier(criterion= 'gini', max_depth = 5, max_features= 15, min_samples_split = 11)
```

```python
dt1.fit(X_train,y_train)
```

```python
y_pred = dt1.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
par = {
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'penalty': ['none', 'l1', 'l2', 'elasticnet'],
    'C':[100, 10, 1.0, 0.1, 0.01]
}
```

```python
tree_cv = RandomizedSearchCV(lr, par, cv = 5,return_train_score=False)
tree_cv.fit(X, y) 
```

```python
print("Tuned Decision Tree Parameters: {}".format(tree_cv.best_params_)) 
print("Best score is {}".format(tree_cv.best_score_))
```

```python
lr = LogisticRegression(solver= 'saga', penalty='l2', C = 1.0)
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
confusion_matrix(y_pred_lr,y_test)
```
